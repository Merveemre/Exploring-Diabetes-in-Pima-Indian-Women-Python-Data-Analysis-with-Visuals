```python
import pandas as pd                                                                                 # Kütüphaneleri içe aktar  
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.width', 1000)                                                           # Görüntüleme ayarlarını belirle
pd.set_option('display.precision', 2)  
data = pd.read_csv('diabetes.csv')                                                                           # Veri setini yükle
print(data.head())                                                                          # Veri setinin ilk 5 satırını yazdır
```

```python
df = pd.read_csv('diabetes.csv')                                     # Veri setini Pandas DataFrame'ine olarak dönüştürerek ata
```

```python
df
```

```python
# 'Pregnancies' ve 'Outcome' dışındaki parametre değerleri '0' olamayacağı için eksik verileri temizleme
columns_to_filter = data.columns.drop(['Pregnancies', 'Outcome'])

for column in columns_to_filter:
    df = df[df[column] != 0]
```

```python
df
```

```python
print(df.describe())                 #Eksik veriler temizlendikten sonra DataFrame'in istatistiksel özetlerini hesapla ve yazdır
```

```python
df.hist(figsize=(7, 7))                                                           # Her bir sütun için histogram grafiği çizme
plt.suptitle('Histogram Charts for Each Column', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
outcome_counts = df['Outcome'].value_counts()                  # 'Outcome' sütununun dağılımını pie chart olarak görselleştirme
colors = ['pink', 'turquoise']
outcome_counts.plot.pie(autopct='%1.1f%%', startangle=90, labels=['Non-Diabetic', 'Diabetic'], colors=colors)
plt.title('Diabetic vs. Non-Diabetic Individuals Ratio')
plt.ylabel('')
plt.show()
```

```python
# lambda fonksiyonu ile 'Outcome' sütununu değerlerine göre sınıflandırma
df['Diabetes_Status'] = df['Outcome'].apply(lambda x: 'Diabetic' if x == 1 else 'Non-Diabetic')

# Yaş gruplarını kategorize etme
df['Age_Group'] = pd.cut(df['Age'], bins=[21, 40, 60, 90], labels=['Young', 'Middle Aged', 'Elderly'])

# BMI gruplarını kategorize etme
df['BMI_Group'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 70], labels=['Underweight', 'Normal Weight', 'Pre-Obese', 'Obese'])
```

```python
df
```

```python
# Diabetes_Status sütununun dağılımı pie chart grafiği
plt.figure(figsize=(3, 3))
df['Diabetes_Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=50, colors=['lightblue', 'pink'])
plt.title('Diabetes Status Distribution')
plt.ylabel('')
plt.show()

# Age_Group sütununun dağılımı pie chart grafiği
plt.figure(figsize=(3, 3))
df['Age_Group'].value_counts().plot.pie(autopct='%1.1f%%', startangle=50, colors=['lightgreen', 'yellow', 'red'])
plt.title('Age Group Distribution')
plt.ylabel('')
plt.show()

# BMI_Group sütununun dağılımı pie chart grafiği
plt.figure(figsize=(4,4))
df['BMI_Group'].value_counts().plot.pie(autopct='%1.1f%%', startangle=50, colors=['pink', 'lightblue', 'green', 'red'])
plt.title('BMI Group Distribution')
plt.ylabel('')
plt.show()
```

```python
# Numerik sütunları seç
numerical_df = df.select_dtypes(include=['number'])

# Numerik verilerle korelasyon matrisini hesaplama
correlation_matrix = numerical_df.corr()

# 'Outcome' ile diğer numerik faktörler arasındaki korelasyonu göster
print("Correlation Analysis - Between Outcome and Other Factors:\n", correlation_matrix['Outcome'])
```

```python
#Korelasyon matrisi heatmap grafiği ile diyabetli olma durumu ile diğer özelliklerin karşılaştırılması
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='Wistia', fmt='.2f')
plt.title('Correlation Heatmap')
plt.xlabel('\n1: Diabetic, 0: Non-Diabetic')
plt.show()
```

```python
plt.figure(figsize=(2, 3))
sns.heatmap(correlation_matrix[['Outcome']], annot=True, cmap='Wistia', fmt='.2f')
plt.title('Outcome Correlation Heatmap')
plt.xlabel('\n1: Diabetic, 0: Non-Diabetic')
plt.show()
```
